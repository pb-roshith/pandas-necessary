```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['a','b','c','d'],
    'age':[20,21,22,23]
})
df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'course':['math', 'english', 'science', 'art']
})
```

```python
merge = pd.merge(df1, df2, how='inner', on='id')
merge
```

```python
merge = pd.merge(df1, df2, how='left', on='id')
merge
```

```python
merge = pd.merge(df1, df2, how='right', on='id')
merge
```

```python
merge = pd.merge(df1, df2, how='outer', on='id')
merge
```

```python
df11 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['a','b','c','d'],
    'age':[20,21,22,23]
})
df22 = pd.DataFrame({
    'course':['math', 'english', 'science', 'art']
})
```

```python
r = pd.concat([df11, df22], axis=1)
r
```

```python
r = pd.concat([df11, df22], axis=0)
r
```

```python
r = pd.concat([df11, df22], axis=0, ignore_index=True)
r
```
